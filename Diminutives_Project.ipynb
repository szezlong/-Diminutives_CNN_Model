```python
import pandas as pd
import numpy as np
from simpletransformers.t5 import T5Model, T5Args
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score

import wandb
import pandas as pd
wandb.login()
```

```python

np.random.seed(47)

df = pd.read_csv("Zdrobnienia.csv", sep=';')

chars_to_remove = "ęóąśłżźćń "

df['podstawowe'] = df['podstawowe'].str.replace(f"[{chars_to_remove}]", '')
df['zdrobnienie'] = df['zdrobnienie'].str.replace(f"[{chars_to_remove}]", '')


df.insert(loc=0, column='prefix', value='')
df = df.rename(columns={"podstawowe": "input_text", "zdrobnienie": "target_text"})

train_df, test = train_test_split(df, test_size=0.3)
eval_df, test_df =  np.array_split(test, 2)

print(len(train_df.index))
print(len(eval_df.index))
print(len(test_df.index))

display(train_df.head())
```

```python


model_type = "t5"
model_name = "allegro/plt5-base"
#special_tokens = ['ą', 'ę', 'ć', 'ł', 'ń', 'ó', 'ś', 'ż', 'ź']

train_args = {
    'evaluate_during_training': True,
    'num_train_epochs': 5,
    'save_eval_checkpoints': False,
    'train_batch_size': 4,
    'eval_batch_size': 4,
    'overwrite_output_dir': True,
    "reprocess_input_data": True,
    "max_seq_length": 128,
    "save_steps": -1,
    "use_multiprocessing": False,
    "fp16": True,
    'wandb_project': "Deminutywy",
    'learning_rate': 3e-4,
    #'special_tokens_list': special_tokens,
}

model = T5Model('t5', model_name, args=train_args, use_cuda=False)


model.train_model(train_df, eval_data=eval_df)
```

```python
podstawowe = test_df["input_text"].values.tolist()
predictions = model.predict(podstawowe)
predictions = [p.replace(' ', '') for p in predictions]

for base, dim in zip(podstawowe, predictions):
    print(f"Diminutive of {base}: {dim}")



```

```python
def remove_pl_chars(strings):
  chars = "ęóąśłżźćń "
  translation_table = str.maketrans('', '', chars)
  return [string.translate(translation_table) for string in strings]

podstawowe = remove_pl_chars(podstawowe)
predictions = remove_pl_chars(predictions)

for base, dim in zip(podstawowe, predictions):
    print(f"Diminutive of {base}: {dim}")
```

```python
f1 = f1_score(y_true=test_df["target_text"], y_pred=predictions, average="macro")
accuracy = accuracy_score(y_true=test_df["target_text"], y_pred=predictions)

print('Accuracy is:', accuracy)
print('F1 measure is:', f1)
```
